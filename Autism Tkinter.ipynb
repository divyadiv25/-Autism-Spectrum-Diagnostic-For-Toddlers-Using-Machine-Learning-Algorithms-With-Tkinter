```python
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Load data and convert 'yes' and 'no' to 1 and 0
data = pd.read_csv('Autism.csv')
data['Jaundice'] = data['Jaundice'].apply(lambda x: 1 if x == 'yes' else 0)
data['Family_mem_with_ASD'] = data['Family_mem_with_ASD'].apply(lambda x: 1 if x == 'yes' else 0)

# Select relevant columns
X = data[['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10', 'Qchat-10-Score', 'Jaundice', 'Family_mem_with_ASD']]
y = data['Class/ASD Traits ']

def perform_lda():
    lda = LinearDiscriminantAnalysis()
    lda.fit(X, y)
    return lda
```

```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

# Load the trained model
lda_model = perform_lda()

# Create a tkinter window
window = tk.Tk()
window.title("ASD Traits Prediction")

# Create tkinter label widgets for the questions
a1_label = tk.Label(window, text="Does your child look at you when you call his/her name?")
a2_label = tk.Label(window, text="How easy is it for you to get eye contact with your child?")
a3_label = tk.Label(window, text="Does your child point to indicate that s/he wants something?")
a4_label = tk.Label(window, text="Does your child point to share an interest with you?")
a5_label = tk.Label(window, text="Does your child pretend? e.g. care for dolls, talk on a toy phone?")
a6_label = tk.Label(window, text="Does your child follow where you are looking?")
a7_label = tk.Label(window, text="If you or someone else in the family is visibly upset, does your child show signs of wanting to comfort them?")
a8_label = tk.Label(window, text="Would you describe your child's first word as:")
a9_label = tk.Label(window, text="Does your child use simple gestures (e.g. wave goodbye)?")
a10_label = tk.Label(window, text="Does your child stare at nothing with no apparent purpose?")
qchat_label = tk.Label(window, text="What is your Qchat-10-Score?")
jaundice_label = tk.Label(window, text="Do you have Jaundice?")
family_mem_label = tk.Label(window, text="Does any of your family member has ASD?")

# Create tkinter entry widgets for user input
a1_entry = tk.Entry(window)
a2_entry = tk.Entry(window)
a3_entry = tk.Entry(window)
a4_entry = tk.Entry(window)
a5_entry = tk.Entry(window)
a6_entry = tk.Entry(window)
a7_entry = tk.Entry(window)
a8_entry = tk.Entry(window)
a9_entry = tk.Entry(window)
a10_entry = tk.Entry(window)
qchat_entry = tk.Entry(window)
jaundice_entry = tk.Entry(window)
family_mem_entry = tk.Entry(window)

# Create a tkinter button to predict the result
def predict_result():
    # Get the user input data from the entry widgets
    new_data = pd.DataFrame({
        'A1': [a1_entry.get()],
        'A2': [a2_entry.get()],
        'A3': [a3_entry.get()],
        'A4': [a4_entry.get()],
        'A5': [a5_entry.get()],
        'A6': [a6_entry.get()],
        'A7': [a7_entry.get()],
        'A8': [a8_entry.get()],
        'A9': [a9_entry.get()],
        'A10': [a10_entry.get()],
        'Qchat-10-Score': [qchat_entry.get()],
        'Jaundice': [jaundice_entry.get()],
        'Family_mem_with_ASD': [family_mem_entry.get()]
    })

    # Convert 'yes' and 'no' to 1 and 0
    new_data['Jaundice'] = new_data['Jaundice'].apply(lambda x: 1 if x == 'yes' else 0)
    new_data['Family_mem_with_ASD'] = new_data['Family_mem_with_ASD'].apply(lambda x: 1 if x == 'yes' else 0)

    # Predict the result for the new input data using the trained model
    prediction = lda_model.predict(new_data)[0]

    # Show the prediction result in a message box
    if prediction == 1:
        messagebox.showinfo("Prediction Result", "Your child is likely to have ASD.")
    else:
        messagebox.showinfo("Prediction Result", "Your child is unlikely to have ASD.")
    
predict_button = tk.Button(window, text="Predict", command=predict_result)

a1_label.grid(row=0, column=0, padx=10, pady=10)
a1_entry.grid(row=0, column=1, padx=10, pady=10)
a2_label.grid(row=1, column=0, padx=10, pady=10)
a2_entry.grid(row=1, column=1, padx=10, pady=10)
a3_label.grid(row=2, column=0, padx=10, pady=10)
a3_entry.grid(row=2, column=1, padx=10, pady=10)
a4_label.grid(row=3, column=0, padx=10, pady=10)
a4_entry.grid(row=3, column=1, padx=10, pady=10)
a5_label.grid(row=4, column=0, padx=10, pady=10)
a5_entry.grid(row=4, column=1, padx=10, pady=10)
a6_label.grid(row=5, column=0, padx=10, pady=10)
a6_entry.grid(row=5, column=1, padx=10, pady=10)
a7_label.grid(row=6, column=0, padx=10, pady=10)
a7_entry.grid(row=6, column=1, padx=10, pady=10)
a8_label.grid(row=7, column=0, padx=10, pady=10)
a8_entry.grid(row=7, column=1, padx=10, pady=10)
a9_label.grid(row=8, column=0, padx=10, pady=10)
a9_entry.grid(row=8, column=1, padx=10, pady=10)
a10_label.grid(row=9, column=0, padx=10, pady=10)
a10_entry.grid(row=9, column=1, padx=10, pady=10)
qchat_label.grid(row=10, column=0, padx=10, pady=10)
qchat_entry.grid(row=10, column=1, padx=10, pady=10)
jaundice_label.grid(row=11, column=0, padx=10, pady=10)
jaundice_entry.grid(row=11, column=1, padx=10, pady=10)
family_mem_label.grid(row=12, column=0, padx=10, pady=10)
family_mem_entry.grid(row=12, column=1, padx=10, pady=10)
predict_button.grid(row=13, column=0, columnspan=2, padx=10, pady=10)

#Run the tkinter event loop
window.mainloop()    
    
```
